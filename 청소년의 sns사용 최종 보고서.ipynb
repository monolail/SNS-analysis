```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')
# 파일 이름 불러오기
print(file_1.keys())
print(file_2.keys())
```

```python
from IPython.display import display

# 업로드한 엑셀 파일을 읽기.
df_xlsx = pd.read_excel(io.BytesIO(file_1['스마트폰_사용여부 (1).xlsx']))

# 엑셀 데이터 전체를 표 형식으로 출력
display(df_xlsx)
```

```python
#불러온 csv 데이터 읽기
df_csv = pd.read_csv(io.BytesIO(file_2['한국언론진흥재단_10대미디어이용통계_SNS 서비스별 이용률 (1).csv']), encoding='euc-kr')

# csv데이터 전체를 표 형식으로 출력
display(df_csv)
```

```python
print(df_csv.columns)
```

```python
# 전체 인원에 대한 SNS별 사용률 그래프
# 막대 그래프 생성
plt.figure(figsize=(10, 6))
sns.barplot(x='내 용', y='전체', data=df_csv, palette='viridis')

# 그래프 제목과 레이블 설정
plt.title('전체 SNS 사용률', fontsize=16)
plt.xlabel('SNS 서비스 종류', fontsize=12)
plt.ylabel('이용률 (%)', fontsize=12)
plt.xticks(rotation=45)

colors = ['skyblue', 'lightgreen', 'gray', 'gold','red','blue']

bars = plt.bar(df_csv['내 용'], df_csv['전체'],color=colors)

# 막대 위에 숫자 표시
for bar in bars:
    yval = bar.get_height()  # 막대의 높이
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# 초등학생에 대한 SNS별 사용률 그래프
plt.figure(figsize=(10, 6))
sns.barplot(x='내 용', y='초등학생', data=df_csv, palette='viridis')


plt.title('초등학생 SNS 사용률', fontsize=16)
plt.xlabel('SNS 서비스 종류', fontsize=12)
plt.ylabel('이용률 (%)', fontsize=12)
plt.xticks(rotation=45)

colors = ['skyblue', 'lightgreen', 'gray', 'gold','red','blue']

bars = plt.bar(df_csv['내 용'], df_csv['초등학생'],color=colors)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
